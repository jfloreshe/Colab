```python
arr = np.random.randint(1,20,10)

matriz = arr.reshape(2,5)

#funciones principales
v = arr.max()
print(v)
v = matriz.max()
print(v)
v = matriz.max(0) # por columna
print(v)
v = matriz.max(1) # por fila
print(v)
v = arr.argmax() # menor indice en todo el array
print(v)
v = matriz.argmax(0) # mpor columna
print(v)
v = matriz.argmax(1) # por fila
print(v)
v = arr.min() # arr.argmin(0), arr.argmin(1), matriz.argmin(0), matriz.argmin(1), matriz.min()
print(v)
v = arr.ptp() #distancia entre el mayor y menor
print(v)
v = matriz.ptp(1)
print(v)
v = np.percentile(arr,50)#percentil
print(v)
arr.sort() # menor a mayor
v = np.median(arr) #mediana
print(arr)
v = np.std(arr) #desviacion estandar
print(v)
v = np.var(arr) #varianza
print(v)
v = np.mean(arr) #media
print(v)
```
